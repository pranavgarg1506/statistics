```python
import seaborn as sns
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
## Generating Binomial Distribution RandomÂ Variable

data = binom.rvs(n = 200, p=0.9,size=40, random_state = 5)
print(data)
```

```python
sns.displot(data,kde =True, color='blue')
plt.show()
```

```python
probability = []
for i in range(0,101):
    temp = round( binom(p=0.60, n = 100).pmf(i),2)
    probability.append(temp)

## plotting a density plot
sns.displot( probability, color = 'blue', kde = True, height = 6, aspect = .7)
plt.show()

## plotting a pmf graph for the variable
sns.displot( x = [ i for i in range(0,101)], y= probability, color = 'blue', kind='hist', height = 6, aspect = .7)
plt.show()
```

```python
## cdf graph

probability = []
total = 0
for i in range(0,101):
    temp = round( binom(p=0.60, n = 100).pmf(i),2)
    total += temp
    probability.append(total)
## plotting a cdf graph for the variable
sns.lineplot( x = [ i for i in range(0,101)], y= probability, color = 'blue')
plt.show()
```

```python
## probabitilty that only 65 students passed the exam is:-

prob = binom(n = 100, p = 0.60).pmf(65)
print(round(prob, 3))
```

```python
## probability of passing at most 65 students

prob = binom(n = 100, p = 0.60).cdf(65)
print(round(prob,2))
```

```python
## probability of passing at least 65 students

prob = binom(n = 100, p = 0.60).cdf(65)
prob = 1 - prob
print(round(prob,2))
```
